#### Problem #2
Solution in Python for comparison to Excel solution.

```python
# constants
L      = 2       # m
Tplate = 343     # K
v      = 1.45    # m/s
Twater = 294     # K
mu     = 9.79e-4 # Pa*s
rho    = 998     # kg/m^3
k      = 0.601   # W/m-K
cp     = 4.18e3  # J/kg-K

# derived quantities
Re = rho*L*v/mu
Pr = mu*cp/k
Nu = 0.332 * Pr**(1.0/3.0) * Re**(1.0/2.0)
h = Nu * k / L
q = h * (Tplate-Twater)

print('Rate of Heat Transfer')
print(str(q)+' W/m^2') 
```

#### Problem #3
Solution in Python for comparison to Excel solution.

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

# calculate initial moles of N2
T = 298.15  # K
P = 1.25    # atm
V = 4000    # L
Rg = 0.0821 # L*atm/mol/K
n_N2 = P*V/(Rg*T) # moles of N2
n_N2 = n_N2 * 0.25  # moles with 25% remaining

# calculate N2 pressure at different V and T
n = 20 # grid points
V = np.linspace(500,1000,n)
T = np.linspace(100,600,n)
# create meshgrid
Vm,Tm = np.meshgrid(V,T)
# calculate pressure
Pm = n_N2 * Rg * Tm / Vm

# plot results
plt.figure()
plt.contourf(Vm,Tm,Pm,cmap='RdBu_r')
plt.colorbar()
CS2 = plt.contour(Vm,Tm,Pm,[1.0,2.5],colors='k')
plt.clabel(CS2, inline=1, fontsize=10)
plt.xlabel('Volume (L)')
plt.ylabel('Temperature (K)')
plt.show() 
```
